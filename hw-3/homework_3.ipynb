# I. Numpy

### Создайте одномерный массив размера 10, заполненный нулями и пятым элемент равным 1. Трансформируйте в двумерный массив.

```python
arr_1 = np.zeros(10, dtype=int)
arr_1[4] = 1
arr_1_resized = np.resize(arr_1, (2, 5))
print(arr_1_resized)
```

### Создайте одномерный массив со значениями от 10 до 49 и разверните его (первый элемент становится последним). Найдите в нем все четные элементы.

```python
arr_2 = np.arange(10, 50)
arr_2_reversed = np.flip(arr_2)
print(arr_2_reversed[arr_2_reversed % 2 == 0])
```

### Создайте двумерный массив 3x3 со значениями от 0 до 8

```python
arr_3 = np.resize(np.arange(9), (3, 3))
print(arr_3)
```

### Создайте массив 4x3x2 со случайными значениями. Найти его минимум и максимум.

```python
arr_4 = np.random.default_rng().random((4, 3, 2))
print(arr_4)
arr_4_max = np.amax(arr_4)
arr_4_min = np.amin(arr_4)
print(f'max = {arr_4_max}')
print(f'min = {arr_4_min}')
```

### Создайте два двумерных массива размерами 6x4 и 4x3 и произведите их матричное умножение. 

```python
rng = np.random.default_rng()
arr_5_a = rng.integers(low=-5, high=5, size=(6, 4))
arr_5_b = rng.integers(low=-5, high=5, size=(4, 3))
arr_5 = np.matmul(arr_5_a, arr_5_b)

print(arr_5_a)
print(arr_5_b)
print(arr_5)
```

### Создайте случайный двумерный массив 7x7, найти у него среднее и стандартное оклонение. Нормализуйте этот массив.

```python
arr_6 = np.random.default_rng().random((7, 7))
print(arr_6)
arr_6_average = np.average(arr_6)
arr_6_std = np.std(arr_6)
print(f'average = {arr_6_average}')
print(f'standard deviation = {arr_6_std}')

# нормализуем по евклидовой норме
arr_6_norm = np.linalg.norm(arr_6)
arr_6_normalized = arr_6 / arr_6_norm
print(arr_6_normalized)
```

### 2. У мужчин средний счёт был на 15% больше, чем у женщин. Мужчины оставляли в среднем на 9% больше чаевых. Плативших мужчин было в 1,8 раз больше, чем плативших женщин.

```python
total_bill_males_mean = df[df['sex'] == 'Male']['total_bill'].mean()
total_bill_females_mean = df[df['sex'] == 'Female']['total_bill'].mean()
print(total_bill_males_mean / total_bill_females_mean)
```

```python
tip_males_mean = df[df['sex'] == 'Male']['tip'].mean()
tip_females_mean = df[df['sex'] == 'Female']['tip'].mean()
print(tip_males_mean / tip_females_mean)
```

```python
males_count = df['sex'][df['sex'] == 'Male'].count()
females_count = df['sex'][df['sex'] == 'Female'].count()
print(males_count / females_count)
```

### 3. В среднем размер счета курящего был на 8% больше, чем у некурящего, однако размер чаевых был примерно одинаковым (меньше 1% разницы). Также, как видно из графика ниже, при большом размере счета некурящие оставляли больше чаевых (в среднем на 31% больше для размера счета выше верхнего квартиля)

```python
df.plot.scatter(x='total_bill', y='tip', c='smoker', cmap="rainbow")
```

```python
total_bill_smoker_mean = df[df['smoker'] == 1]['total_bill'].mean()
total_bill_nonsmoker_mean = df[df['smoker'] == 0]['total_bill'].mean()
print(total_bill_smoker_mean / total_bill_nonsmoker_mean)
```

```python
tip_smoker_mean = df[df['smoker'] == 1]['tip'].mean()
tip_nonsmoker_mean = df[df['smoker'] == 0]['tip'].mean()
print(tip_smoker_mean / tip_nonsmoker_mean)
```

```python
cond_big_bill = (df['total_bill'] > df['total_bill'].quantile(0.75))
tip_smoker_big_bill_mean = df[(df['smoker'] == 1) & cond_big_bill]['tip'].mean()
tip_nonsmoker_big_bill_mean = df[(df['smoker'] == 0) & cond_big_bill]['tip'].mean()

print(tip_nonsmoker_big_bill_mean / tip_smoker_big_bill_mean)
```

### 4. На графиках ниже видно, что средний счёт и чаевые были немного выше в субботу и воскресенье, чем в четверг и пятницу. В пятницу было гораздо больше курящих, чем некурящих. В субботу и воскресенье плативших мужчин было в процентном выражении ощутимо больше, чем в четверг и пятницу. Также в пятницу было гораздо меньше счетов, чем в остальные дни. 

```python
df_copy = df.copy()
df_copy['male'] = pd.to_numeric(df_copy.sex.map({'Male': 1, 'Female': 0}))
df_mean_with_count = df_copy.groupby('day') \
    .agg(count=('day', 'size'), \
         mean_total_bill=('total_bill', 'mean'), \
         mean_tip=('tip', 'mean'), \
         smoker_ratio=('smoker', 'mean'), \
         male_ratio=('male', 'mean'))
df_mean_with_count
```
